```python
import matplotlib.pyplot as plt
```

```python
x= [1,2,4,5]
y= [1,2,3,5]
plt.plot(x,y, marker= 'o', color='b', linestyle='--')
plt.title('Basic line plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
plt.show()
```

#** 2. Figure and** **Axes**

```python
fig, ax = plt.subplots()

ax.plot(x,y, marker= '.', linestyle='--', color= 'r')
ax.set_title('Figure and Axes')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
```

# **3. Multiple Plots in One Figure (Subplots)**

```python
#Data
a= [0,1,2,3,4,5]
b= [0,1,4,9,16,25]
b1= [25, 20, 15, 10, 5, 0]
#Create a figure with two subplots
#1= Row , 2= column means in one row 2 columned graph will be drawn;
#figsize= 10 inches wide and 4 inches tall.
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,4))
#First Subplot
ax1.plot(a,b, color='g', marker='o')
ax1.set_title('Quadratic Growth')
#Second subplot
ax2.plot(a,b1, color='r', marker='o')
ax2.set_title('Liner Decline')
#show the plot
plt.tight_layout()
plt.show()
```

# **4. Bar Charts**

```python
catergo = ['A','B','C','D']
values= [4,7,1,8]
#Bar chart
plt.bar(catergo, values, color=['blue','orange','green','red'])
plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')

plt.show()
```

## **5. Scatter Plot**

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 8]
sizes = [20, 50, 80, 200, 500]  # Varying marker sizes

# Scatter plot
plt.scatter(x, y, c='red', s=sizes, alpha=0.5)  # s is for size, alpha is for transparency
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```

# **6. Histograms**

```python
import numpy as np

# Random data
data = np.random.randn(1000)

# Histogram
plt.hist(data, bins=30, color='purple', alpha=0.7)
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

# **7. Customizing Plots**

```python
# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Customizing the plot
plt.plot(x, y, color='teal', linestyle='-.', linewidth=2, marker='o', markerfacecolor='red')
plt.title('Customized Plot')
plt.xlabel('X-axis')
plt.ylabel('Sine Wave')
plt.grid(True)  # Adds grid lines

# Show the plot
plt.show()
```

# **8. Saving Plots**


```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Plot
plt.plot(x, y, marker='s', color='brown')
plt.title('Saving Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Save the plot to a file
plt.savefig('plot_example.png')  # Saves the figure as a PNG file

# Show the plot
plt.show()
```
